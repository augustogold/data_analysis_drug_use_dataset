### 1.1 Load the file with the `csv` module and put it in a Python dictionary

The dictionary format for data will be the column names as key, and the data under each column as the values.

Toy example:
```python
data = {
    'column1':[0,1,2,3],
    'column2':['a','b','c','d']
    }
```

````python
file = open(r"C:\Users\augus\Documents\Study\General Assembly\Data Science\DSI22-lessons\projects\project-2\sat_scores.csv")

csvreader = csv.reader(file)

header = next(csvreader)

print(header)

rows = []

for row in csvreader:
    rows.append(row)

print(rows)

file.close()
````

````python
#Creating a dict from the header and rows variables

def file_csv_into_dict(header, rows):
    
    new_dict = {}
    
    columns_0 = []
    columns_1 = []
    columns_2 = []
    columns_3 = []
    
    for i in rows:
        
        for j, z in enumerate(i):
            
            if j==0:
                columns_0.append(z)
                
            elif j==1:
                columns_1.append(z)
                
            elif j==2:
                columns_2.append(z)
                
            elif j==3:
                columns_3.append(z)
            
    for t in range(len(header)):
        
        if t==0:
            new_dict[header[t]] = columns_0
        
        elif t==1:
            new_dict[header[t]] = columns_1 
        
        elif t==2:
            new_dict[header[t]] = columns_2
            
        elif t==3:
            new_dict[header[t]] = columns_3
            
    
    return new_dict
````

````python
SAT_dict = file_csv_into_dict(header, rows)
print(SAT_dict)
````

### 6.1 Create the correlation matrix of your variables (excluding `State`).

What does the correlation matrix tell you?


````python
#To create the correlation matrix, first I need to procede a 'standardization' of the variables(columns).

#Before everything, i'll create a new df to not change the values of the first one

df_standart = SAT_df.copy()

#Rate

df_standart['Rate'] = (df_standart['Rate'] - df_standart['Rate'].mean()) / df_standart['Rate'].std()

#Verbal

df_standart['Verbal'] = (df_standart['Verbal'] - df_standart['Verbal'].mean()) / df_standart['Verbal'].std()

#Math

df_standart['Math'] = (df_standart['Math'] - df_standart['Math'].mean()) / df_standart['Math'].std()

#(let's drop those last two columns to make things prettier)

df_standart.drop(columns=['Verbal - Math', 'Math - Verbal'], inplace=True)
SAT_df.drop(columns=['Verbal - Math', 'Math - Verbal'], inplace=True)

#the result nice and clean 

df_standart
````

````python
#Now, lets see the proper correlation

df_standart[['Rate', 'Verbal', 'Math']].corr()
````

````python
#Shall we see with the help of visuals?

import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
sns.heatmap(df_standart[['Rate', 'Verbal', 'Math']].corr(), annot=True)
plt.show()
````

What does the correlation matrix tell you?

    The numbers (and the plot) show very clearly that all variables (columns) are almost perfect related to each other (at least numerically speaking)! Worth noting that, more evidence would be necessary to infer that this correlation actually exists in real life.
